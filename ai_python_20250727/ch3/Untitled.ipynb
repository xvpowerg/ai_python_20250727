```python
txt = input("")
strList = []
strList.append(txt)
msg = txt
count = 0
inputMsg = f"-{txt[-1]}-"
while count < 5:
    tmpTxt = input(inputMsg)
    if tmpTxt[0] == txt[-1]:
        txt = tmpTxt
        strList.append(tmpTxt)
        msg = "-".join(strList)
        inputMsg = f"-{txt[-1]}-"
    else:
        inputMsg = f"-{txt[-1]}-"
        count += 1
    print(":",msg)    
```

## Filter


```python
myList = list(range(1,21))
print(myList)
def myFilter(data):
    result = []
    for d in data:
        if d % 2 == 0:
            result.append(d)    
    return result
newList = myFilter(myList)    
print(newList)
```

```python
myList = list(range(1,21))
print(myList)
def myFilter(data,func):
    result = []
    for d in data:
        if func(d):
            result.append(d)    
    return result
newList = myFilter(myList,lambda x : x %2 == 0)    
print(newList)
newList2 = myFilter(myList,lambda x : x %3 == 0)   
print(newList2)
```

```python
myList = list(range(1,21))
print(myList)
filterObj = filter(lambda x:x%3 ==0, myList)
print(filterObj)
print(list(filterObj))
```
