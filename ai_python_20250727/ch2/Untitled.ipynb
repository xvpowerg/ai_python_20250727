## 字串區間


```python
str2 = "ABCDEFGHI"
print(str2[1])
print(str2[7])
print(str2[-2])

print(str2[:4])#由頭到小於4的位置取值
print(str2[2:])#由2到尾部的位置取值

print(str2[1:4])#由1到小於4的位置取值
print(str2[-4:-1])#由-4到不包含-1的位置
#DEF
print(str2[3:-3])
print(str2[3:6])
print(str2[::2])#ACEGI 每次index(編號)+2

print(str2[::-1])
print(str2[::-2])

print(str2[-2:-4:-1])
print(str2[-3:-7:-1])
print(str2[5:2:-1])
```

# List VS Tuple
+ tuple 與 list 的差異
  + 括號不同：
    + tuple 是用小括號將元素包起來。
    + list 是用中括號將元素包起來。
+ 元素可否變更：
   + tuple 決定了元素的內容就不可以變更。
   + list 
## list 介紹
+ list 內資料用逗號隔開，放置於中括號的範圍內。
  + list_name = [1,2,3 ]
+ list 可以儲存不同類型的值或項目，如數字和字串。
  + list_name = [1, 2, ‘Python’, ‘lists’ ] 
+ list 內的資料可以進行增加、更新與移除。
+ 可透過.index( )方法取得指定資料的索引。可以隨時修改

```python
myList = [1,"Ken",99]
myTuple = ("Vivin",18,166.2) #唯讀
print(myList[1])
print(myTuple[0])
print(myTuple[-1])

myList[-1] = 120
print(myList[-1])
myTuple[-1] = 156.3#唯讀不可改會錯
```

```python
myList = [18]
print(myList,type(myList))
myTuple = (25)#tuple只有一個數值的時候要小心
print(myTuple,type(myTuple))
myTuple2 = (83,)#tuple
print(myTuple2,type(myTuple2))
```

```python
list1 = ["Ken","Vivin","Joy","Lucy"]
print(list1[:])
print(list1[:3])
print(list1[::-1])
print(list1[2::-1])
print(list1[1])
print(list1[1:-1])
```

```python
list1 = ["A","B","C","D"]
newList = list1[1:3]
print(list1)
print(newList)#會產生全新的List
print("===================")
newList[0] = 88
print(list1)
print(newList)
```

## 移除 
+ list 可透過以下三種方式移除項目：
  + list.remove( )
  + list.pop( )
  + del list[ ]

###  remove

依據內容進行刪除

list.remove(項目內容)

```python
list1 = ['1','x','2','x','3']#由左往右移除
print(list1)
list1.remove('x')
print(list1)
list1.remove('x')
print(list1)
#list1.remove('x')#x not in list 
```

```python
list1 = ['1','x','2','x','3']#由左往右移除
key = 'x'
if key in list1:
    print(key,"在list")
else:
    print(key,"不在list")
```

```python
list1 = ['1','x','2','x','3']#由左往右移除
key = "x"
while key in list1:
    list1.remove(key)
print(list1)    
    
```

### pop

如果 pop( ) 內沒有參數，則移除最後一筆資料。

pop() 內若有參數，則移除指定位置的資料。

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lindy")
list2.append("Lucy")
print(list2)
print(list2.pop())#彈出尾部
print(list2)
print(list2.pop(0))#彈出尾部
print(list2)
print(list2.pop())#彈出尾部
print(list2)
print(list2.pop())#彈出尾部
print(list2)
print(list2.pop())#彈出尾部
```

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lindy")
list2.append("Lucy")
print(list2)

while len(list2) > 0:
    print(list2.pop())


```

### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list2 = ["A","B","C","D","E"]
print(list2)
del list2[2]
print(list2)
del list2[1:3]
print(list2)
```

## 字串的Join

## join

join( ) : 於字串或著字元之間加入指定的文字

指定文字.join(要插入的字串)

```python
nameList = ["Ken","Vivin","Lindy"]
#"Ken,Vivin,Lindy"
msg = ""
for i in range(len(nameList)):
    msg +=nameList[i]
    if i < len(nameList)-1:
        msg += ","

print(msg)
```

```python
nameList = ["Ken","Vivin","Lindy"]
msg = ",".join(nameList)#join的內容一定是字串
print(msg)
```
