### 使用 loc(文字 m:n 會包含到n)
+ "location" 的縮寫
+ .loc[ ], 中括弧裡面是先 row 後 column，以逗號分開。
+ row 與 column 要放入的是 row 標籤與 column 標籤。
+ 可於 row 或 column 加入「：」 ，代表指定起訖範圍。

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(48).reshape(6,8),
                  index = ["a","b","c","d","e","f"],
                 columns = ["s","t","u","v","w","x","y","z"])
print(df)
print()
print(df.loc["b",:])
print("1----")
print(df.loc["c","w"])
print("2----")
print(df.loc[['c','e'],['w','y']])
print("3----")
print(df.loc['c':'e',"w":"y"])
print("4----")
print(df.loc[:,"x"])
```

### 使用 iloc
+ iloc[ ] 與 loc 一樣，中括弧裡面也是先 row 後 column，行列標籤用逗號分開。
+ 與 loc 不同的之處是，.iloc 是根據索引來進行選擇。

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.arange(48).reshape(6,8),
                  index = ["a","b","c","d","e","f"],
                 columns = ["s","t","u","v","w","x","y","z"])
print(df)
print()
print(df.iloc[1,:])
print("1----")
print(df.iloc[2,4])
print("2----")
print(df.iloc[[2,4],[4,6]])
print("3----")
print(df.iloc[2:4,4:6])
print("4----")
print(df.iloc[:,-3])
```
