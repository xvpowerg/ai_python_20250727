```python
import pandas as pd
data = {'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'year': [2012, 2012, 2013, 2014, 2014],
        'reports': [4, 24, 31, 2, 3],
        'coverage': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data, index = ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'])
print(df)
print("="*50)
##coverage 大於50的
print(df[df["coverage"] > 50])
print("="*50)
print(df.query("coverage > 50"))
```

```python
import pandas as pd
data = {'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'year': [2012, 2012, 2013, 2014, 2014],
        'reports': [4, 24, 31, 2, 3],
        'coverage': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data, index = ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'])
print(df)
print("="*50)
print(df[(df["coverage"] > 50) & (df["reports"] < 4)  ])

v1 = 50
v2 = 4
print("="*50)
print(df.query("coverage > @v1 & reports < @v2"))
```

```python
import pandas as pd
import numpy as np
df1 = pd.DataFrame(np.random.randn(10,2),index=[1,4,6,2,3,5,9,8,0,7],columns = ['col2','col1'])
print(df1)
df2 = df1.sort_index()
print("="*10)
print(df2)
print("="*10)
df3 = df1.sort_index(ascending=False)
print(df3)
print("="*10)
df4 = df1.sort_index(axis=1)
print(df4)
```
