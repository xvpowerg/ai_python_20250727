## break 代表不回頭，跳出迴圈。
## continue 代表繼續執行下一個迴圈。

```python
#1~10
for i in range(1,11):
    print(i,end = " ")
    if i == 5:
        break #停止迴圈
```

```python
for i in range(1,11):
    if i == 5 or i == 8 :
        continue
    print(i,end= " ")
```

```python
print("請輸入大於80的整數")
while True:
    a = int(input("請輸入數字"))
    if a > 80:
        print("a=",a)
        break
        
```
